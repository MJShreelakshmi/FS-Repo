#### 1. What is Python? What are the benefits of using Python?
        Answer: Python is an interpreted, high-level, and general-purpose programming language. Benefits include easy syntax, readability, extensive libraries and frameworks, community support, and versatility for different types of programming such as web development, data analysis, artificial intelligence, and more.

#### 2. How is Python interpreted?
        Answer: Python is an interpreted language, meaning its code is executed line-by-line, which makes debugging easier and allows for more interactive coding sessions. The Python interpreter reads the code, translates it into machine language, and then executes it.

#### 3. What are Python's key features?
        Answer: Python features include simplicity and readability, extensive standard libraries, cross-platform compatibility, support for both procedural and object-oriented programming, and dynamic typing.

#### 4. Explain the difference between list, tuple, and set.
        Answer:
            List: Ordered, mutable, and allows duplicate elements.
            Tuple: Ordered, immutable, and allows duplicate elements.
            Set: Unordered, mutable, and does not allow duplicate elements.

#### 5. What is a dictionary in Python? How do you create one?
        Answer: A dictionary is a collection of key-value pairs. It is unordered, mutable, and indexed by keys. You can create a dictionary using curly braces {} or the dict() function. Example:

    my_dict = {'name': 'Alice', 'age': 25}

#### 6. How does Python manage memory?

    Answer: Python uses a private heap space for memory management. The Python memory manager handles the allocation and deallocation of memory for data structures. It also includes a built-in garbage collector, which recycles all the unused memory to make it available for heap space.

What are Python decorators?

    Answer: Decorators are a way to modify the behavior of a function or class. They are typically used to add functionality to existing code in a reusable way. You define a decorator as a function that takes another function and extends its behavior without explicitly modifying it. Example:

    def my_decorator(func):
        def wrapper():
            print("Something is happening before the function is called.")
            func()
            print("Something is happening after the function is called.")
        return wrapper

    @my_decorator
    def say_hello():
        print("Hello!")

    say_hello()

Explain list comprehensions with an example.

    Answer: List comprehensions provide a concise way to create lists. They consist of brackets containing an expression followed by a for clause, and can include optional if clauses. Example:

    squares = [x**2 for x in range(10) if x % 2 == 0]
    # Output: [0, 4, 16, 36, 64]

What are generators in Python?

    Answer: Generators are a simple way to create iterators using a function that yields values one at a time. They allow you to iterate over data without storing the entire data set in memory. Example:

        def my_generator():
            for i in range(10):
                yield i

        for value in my_generator():
            print(value)

Expert Questions

    What is the Global Interpreter Lock (GIL) in Python?
        Answer: The GIL is a mutex that protects access to Python objects, preventing multiple threads from executing Python bytecodes at once. This ensures thread safety but can be a bottleneck in CPU-bound and multithreaded code. However, it doesn't affect performance in I/O-bound and multiprocessing applications.

    How do you handle exceptions in Python?
        Answer: Exceptions in Python are handled using try and except blocks. You can catch specific exceptions or handle all exceptions using a generic except block. Example:

    try:
        result = 10 / 0
    except ZeroDivisionError:
        print("Cannot divide by zero.")
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        print("This will execute no matter what.")

Explain the concept of metaclasses in Python.

    Answer: Metaclasses are classes of classes that define how classes behave. A class is an instance of a metaclass. Metaclasses allow you to control the creation and behavior of classes, including how attributes and methods are defined. Example:

        class MyMeta(type):
            def __new__(cls, name, bases, dct):
                print(f"Creating class {name}")
                return super().__new__(cls, name, bases, dct)

        class MyClass(metaclass=MyMeta):
            pass

        # Output: Creating class MyClass

Practical Questions

    How do you perform file I/O in Python?
        Answer: File I/O in Python is done using the open() function, which returns a file object. You can read from and write to the file using methods like read(), write(), and close(). Example:

    with open('example.txt', 'w') as file:
        file.write('Hello, world!')

    with open('example.txt', 'r') as file:
        content = file.read()
        print(content)

What is the difference between deepcopy and shallowcopy?

    Answer: A shallow copy creates a new object but inserts references into it to the objects found in the original. A deep copy creates a new object and recursively copies all objects found in the original. Example:

        import copy

        original = [1, [2, 3], 4]
        shallow = copy.copy(original)
        deep = copy.deepcopy(original)

        original[1][0] = 'changed'
        print(shallow)  # Output: [1, ['changed', 3], 4]
        print(deep)     # Output: [1, [2, 3], 4]
