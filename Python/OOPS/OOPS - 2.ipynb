```python
class Player:
    def __init__(self, pn, pa, pc, pm, pw, pr):
        self.pname = pn
        self.page = pa
        self.country = pc
        self.pmatches = pm
        self.pwickets = pw
        self.pruns = pr
        
class Team:
    def __init__(self, Nop):
        self.no_of_palyers = Nop
        self.pl = []
        for i in range(self.no_of_palyers):
            pn = input('Enter Pname: ')
            pa = input('Enter Page: ')
            pc = input('Enter country: ')
            pm = int(input('Enter number of matches played: '))
            pw = int(input('Enter number of wickets taken: '))
            pr = int(input('Enter no of runs scored: '))
            po = Player(pn, pa, pc, pm, pw, pr)
            self.pl.append(po)
            
    def max_wkt_taker(self):
        mwto = self.pl[0]  # max_wicket_taker_object 
        for po in self.pl:
            if po.pwickets>mwto.pwickets:
                mwto = po
        return mwto.pname
    
    def max_run_scorer(self):
        mrs = self.pl[0]
        for po in self.pl:
            if po.pruns> mrs.pruns:
                mrs = po
        return mrs.pname
```

### Inheritence
1. Inheritence represents Is A Relaitonship between entities
2. Inheritence is the process of accessing the properties of one class into another
3. The class from which the properties are acquired are known as Base/Parent/Super Class
4. The class into which the properties are acwuired are known as Derived/Child/Sub Class

Note:
1. All the parent class properies can be used by child
2. But none of the child class properties can be used by the parent Class

#### Types of Inheritence
1. Single-level Inheritence (One Parent- One Child)
2. Multiple Inheritence ( Two Parent - one Child)
3. Hierarchial Inheritence (Single Parent - Multiple Child)
4. Multi-Level Inheritence (One Child - it's Child (another) - Inheritence in multiple levels)
5. Hybrid Inheritence (Combination of more than One type of Inheritence)

#### Single level Inheritence
1. In this type of Inheritence, only one Child is Inherited from One parent
2. Parent class properties can be accessed in the Child class
3. But Child class properties cannot be accessed in the Parent class

While creating Child class dictionary, first the properties of Parent class are derived then the Child class properties are derived into that dictionary.

Syntax:

        class Parent:
            pass
            
            
        class Child(Parent):
            pass
            
Note:
1. Direction of Inheriting the properties is from Right to Left
2. Direction of Execution of the properties is from Left to Right

```python
class Vinay:
    house='1BHK'
    bike='Chetak Bike'
    money=1000000
    
lakshmi = Vinay()

class Narayana(Vinay):
    bike='KTM390'
    car='KIA'
    
asha=Narayana()
```

In case of Inheritence, while defining the methods and properties of the child class, the following rules has to be followed:
1. Property with same value in parent and child class --> Do not do anything
2. Property with different value in parent and child class --> Redefine the property with new value in child class
3. New property in child class --> Create a new variable
4. Method with same implementation in parent and child class --> Do not do anything
5. Method ith different impleemntation in parent and child class --> Redefine the method (Method Overriding)
6. Ne Method creation --> Create a new variable

#### Polymorphism
* Polymorphism is a process of same method having different functionalities (with different objects)
* We can achieve polymorphism in 2 ways:
1. Method Overriding
2. Method Overloading
#### Method Overriding
* In case of Inheritence, we perform Method Overriding
1. It is the process of changing the implementation of Super (Parent) class method in Sub class (Child class)
2. To override the Super class method, we have to implement (redefine) the method with the same name in Sub (Child) class as well
3. We can override _ _init_ _,static methods, class methods and object methods as well

Example:

```python
class Animal:
    def move(self):
        print('Animal is Running')
        
class Snake:
    def move(self):
        print('Snake is Crawling')
        
S=Snake()
S.move()
```

#### Note:
If we perform only Method Overriding, code redundancy will increase. So inorder to avoid that we need to perform Chaining

#### Chaining:
It is the process of calling the constructor or methodof Constructor or Method of Another Class in Current Class / Constructor

We have 2 types of chaining:
1. Constructor Chaining
2. Method Chaining

#### Constructor Chaining
It is the process of calling the constructor of another class in current class constructor

We can perform it in 2 ways:
1. Using Super method
2. Using Class Name

##### Super() method
1. Super() method is used only in case of Inheritence
Syntax:

    super().__init__(arguments)
    
##### Class Name
Calling of Constructor by using Class Name can be used incase of Inheritence and Non-inheritence as well

Syntax:
    
    class_name.__init__(self, arguments)
    
#### Method Chaining
It is the process of Calling the Method of Another Class in Current Class Method

We can perform it in 2 ways:
1. Super Method
2. By Class Name

##### Super() Method
Super() method is used only in case of Inheritence

Syntax:

    super().method_name(argumets)
    
##### Class Name
Calling of method by using class can be used in case of Inheritence as well as Non-Inheritence

Syntax of calling Object Method by Class:

    class_name.method_name(self, arguments)
    
Syntax for calling Class Methods by chaining Class:

    class_name.method_name(arguments)
