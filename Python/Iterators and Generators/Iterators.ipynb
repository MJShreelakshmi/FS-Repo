#### Iterators
1. Iteration is a process of travsersing through the sequence of and extracting a value one after the other
2. The operators which do the process of iteration is known as Iterators
3. For loop works on the principle of iterators
4. Strings, lists, tuples, dictionaries and sets are all iterables

Iterator works with 2 set of operations:
1. Initialization: 
    
        1. It creates an iterable object from a gien iterable
        2. Creates one variable and assigns first vlue of iterable object
        3. It returns iterable object for traversing
2. Traversing:
    
        1. Extract each and every element from iterable object
        2. When there is no element left to extract, create a StopIteration Error

#### Classification
1. Built-in Iterators
2. Custom/User-defined Iterators

        In case of built-in Iterators, iter performs initilization
        It creates and returns Iterable object

#### Custome Iterator
1. We have to create a class with _ _iter_ _() and _ _next_ _() methods to createa custome iterator / User-defined Iterators
2. _ _iter_ _() is used for initializing the temporary variable with starting value and used for returning the iterable object
3. _ _next_ _() is used for returning the next item in the sequence. On reaching the end and in subsequent calls, it must raise StopIteration Error

#### Demonstration of range function by using custom iterator

```python
class CIRC():
    def __init__(self, sv, ev, up=1):
        self.sv = sv
        self.ev = ev
        self.updation = up
        
    def __iter__(self):
        print('__iter__ is called')
        self.i = self.sv
        return self
    
    def __next__(self):
        print('__next__ is called')
        if self.i < self.ev:
            res = self.i
            self.i += self.updation
            return res
        else:
            raise StopIteration
            
CIO = CIRC(1,5)
for i in CIO:
    print(i)
```

```python
# create custom iterator to get the cubes of gien range
class Cubes():
    def __init__(self, sv, ev, up=1):
        self.sv = sv
        self.ev = ev
        self.updation = up
        
    def __iter__(self):
        self.i = self.sv
        return self
    
    def __next__(self):
        if self.i < self.ev:
            res = self.i**3
            self.i += self.updation
            return res
        else:
            raise StopIteration
            
co = Cubes(1,5)
for i in co:
    print(i)
```

```python
# create custom iterator which generates a Fibonacci series

class fibo:
    def __init__(self, fv, sv, n):
        self.fv = fv
        self.sv = sv
        self.n = n
        
    def __iter__(self):
        self.i = 1
        return self
    
    def __next__(self):
        if self.i <= self.n:
            self.i += 1
            res = self.fv
            self.fv = self.sv
            self.sv = self.fv + res
            return res
        else:
            raise StopIteration
ob = fibo(2,3,10)

for i in ob:
    print(i, end=' ')
```
