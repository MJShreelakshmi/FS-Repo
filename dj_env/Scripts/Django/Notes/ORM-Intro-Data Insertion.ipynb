### ORM


#### N-Tier Architecture
In this type of architecture, we will have more than 2 layers. i.e., Our application will have following ends or phases
1. FRONT-END
2. BACK-END
3. DATABASE END

In developement lingo, we call the database end as Model layer
#### Models
Popularly we can store the data in 2 formats
1. Tables
2. Document Formats
In Djnago frameworks, we will store the data in the form of tables.

As it is relational DBMS, we hace to use languages like MySQL, PostgreSQL, SQL+ etc to manipulate the data.

Since it is difficult to learn(or shift the DBs as per the reuirement) django has come up with the concept called ORM to manipulate the data.

#### ORM (Object Relational  Modelling)
It is a concept which is repsonsible for writing database queries in the form of Python Object Oriented Programming.

Advantages of ORM are:
1. We can use any type of database
2. By default, django is provided with Sqlite3 database
3. We can connect other databases instead of Sqlite3 just by pip installing it and making some changes in the database section of our settings.py

#### Creation of tables in database 
In order to create a table using ORM, we have to follow below considerations

Considering
1. Class in OMR ----> Table in database
2. Class variable in ORM  ----> A column in database
3. Object in ORM ----> A row of data in a database

#### Steps to create Models
1. Go to models.py in the app
2. Each and every model class must be inherited from Model class as shown below
3. Create a class which represents a table name as table_name with the syntax below

       class table/model_name(models.Model):
           class_member = models.fieldname(arguments)

Here, 
* The class_member is used to represent column_name
* fieldname is used to represent datatype
* Arguments act as constraints
We have to create a method with a name _ _str_ _, which is used for returning a string value.            

#### Contents of models.py

        from django.db import models
        # create your models here.

        class Topic(models.Model):
            topic_name = models.CharField(max_length =30, primary_key = True)
            def __str__(self):
                return self.topic_name

        class Webpage(models.Model):
            topic_name = models.ForeignKey(Topic, on_delete = models.CASCADE)
            name = models.CharField(max_lengtyh = 100)
            url = models.URLField()

        class Access_Records(models.Model):
            name = models.ForeignKey(Webpage, on_delete = models.CASCADE)
            date = models.DateField(auto_now = True)
            author = models.CharField(max_length = True)

#### Operations that can be done with tables (CRUD Operations)
1. Insertion of data
2. Retrieving of data
3. Deleting the data
4. Updating the data
#### 1. Insertion of data
We can insert the data into our models in 3 ways:

    1. Creating an object and passing the values as key, values
    2. By create method
    3. By get_or_create method

#### 1. Creating an object to insert data
syntax:

    object_name = Class(col_1 = val_1, col_2 = val_2...)

Inorder to save the inserted data, we have to make use of save() method of Model class

#### Changing the cmd into interactive shell
Run the following command to enter into interactive console

    python manage.py shell
The command below is used to exit from the interactive console

    exit()
#### 2. Inserting data by using create method
1. It is exclusively used for creating new object
2. If the object is not present in the database, it will create a new object otherwise it will give an error
3. Here, unlike the previous method of direct creation of object, we need not use save method to save the newly inserted data

syntax:

    object_name = class_name.objects.create(col_1 = val_1, col_2 = va_2, ...)
When we use create method to insert the data, it returns the instance directly
#### Note:
While inserting data into ForeignKey column, one should provide its Parent Table object, not the value itself
#### 3. Inserting data by using get_or_create method
1.get_or_create method is used for creating object and returning output in the form of tuple
2. save() method is not necessary to save it to the database if we use get_or_create method to create an object

syntax:

    object_name = class_name.objects.get_or_create(col_1 = val_1, col_2 = val_2,... )[0]
* This method saves the data automatically
* This method returns a tuple with (instance, boolean_flag)
#### Note:
get_or_create method is a universal method used to insert data in real time
