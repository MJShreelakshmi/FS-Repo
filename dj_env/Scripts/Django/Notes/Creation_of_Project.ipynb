### Virtual Environment
Virtual Environment is used for dividing a system into number of isolated brand new systems

Purpose of creating isolated parts is to avoid affecting of the downloaded python package versions to the entire system

#### Process to create a virtual environment
1. Install virtalenv package with following command

           pip install virtaulenv
2. Create a virtaul environment by using below syntax

           virtualenv venv_name
                   or
           pyhton -m virtualenv venv_name (for an older system)
3. Immediately after creating virtaulenv, a brand new system will be created
4. This sytsem will have following files

           1. Lib 2. Scripts 3. Environment Configurations
   
6. Enter into the created virtual environment

7. Activate the virtual environment

           cd scripts
           activate
   
8. Installing Django within the virtual environment (Since  the django software version should not be affecting the entire system )
9. To turn off the active virtual environment after use, run the following command

           deactivate

#### Creation of Django Project
Project is a container which is consisting of many features or applications
#### Procedure to create a Project
1. Syntax for creating a project in Django

           django-admin startproject project_name
2. Once the above command is executed, a container will be created under the name of the project i.e., say project_name. Within the container, we will have 2 files.

               1. project_name folder
                  This is a folder with the same name as the project name and is responsible for integrating the application
               2. manage.py
                  It is a managerial file which is responsible for controlling and triggering (starting and stopping) of the server
#### Information of files present inside the project folder
#### 1. _ _ init_ _.py
It is a blank python file which is used for making or converting a folder with collection of modules into a package
#### 2. asgi.py (Aynchronous Server Gateway Interface )
asgi.py is used for creating standard interface (Platform where the Servers interact with each other) between Framework, Project and Webserver    
#### 3. settings.py
It is the file where we will be making all the changes with respect to our project, application, database, static files and templates etc.
#### 4. urls.py
This is responsible for performing the process of url mapping

    Url mapping: It is the process of connecting or mapping the url with their respective function or class
#### 5. wsgi.py (Web Server Gateway Interface)
wsgi.py is used for hosting our application or software into the web.


#### Creation of Application 
Application is the feature of a project. A single project can have multiple applications

#### Creation of Application
Syntax for creating an application

        python manage.py startapp app_name
#### Information of the files within the application 
#### 1. _ _ init_ _.py
It is a blank file which is responsible for representing a folder as a package
#### 2. apps.py
It is a file where we perform all the application related configurations\
#### 3. models.py
It is a file where we create our models (tables)
#### 4. admin.py
It is a file responsible for registering the created models
#### 5. Migrations
It is used for storing the information related to models (tables) of the database
#### 6. views.py
It is a file where we write views (functions or classes) which are responsible for responding to the received requests
#### 7. tests.py
It is a file where we write our test cases which is responsible for testing whether the created functions or classes are working fine or not

#### URL : Uniform Resouce Locator
#### Parts of URL

       http://127.0.0.1:8000/suffix

Url is a combination of certain things mentioned below:

       protocol://BaseAddress:Port-number/suffix

#### 1. Protocol
It is used for specifying the standard rules that needs to be followed for data processing and communication
#### 2a. http
* It is an unsecured and open-source protocol
* We cannot process large amount of data
* Used only while developing
#### 2b. https
* It is the secured and paid protocol
* We can process large amount of data
* Used while developing and after developing as well
#### 3. baseurl
* It is the logical address of the server
* As we run the server within our system, django creates a virtual address i.e., 127.0.0.1
#### 4.port number
* It is the address of the channel through which the control enters the server
* Default django port number (channel address) is 8000
#### 5. suffix
* Suffix is used to navigate to a particular page from multiple pages
* There are 2 types of suffixes

      1. primary suffix
      2. secondary suffix

#### Views
* Views are the callables which take request and return a response in the form of http
* We can return either string or HTML file as response

#### Procedure to connect app to the project and create urls, views

1. create a project
2. create an application with command:    
    
        python manage.py startapp app_name
3. Register the app with the project
   
        * go to settings.py of project
        * scroll down to find INSTALLED_APPS variable. Register the app by specifying the name as element of the list
4. Create views. There are 2 types of views (1) function based views  (2) class based views

        * Go to settings.py of application
        * Create a function which accepts the request as first argument. 
        * To respond in the string form, import HttpResponse function from django.http

   Example:
        
            def function_name (request):
                return HttpResponse('Data in the form of response')
6. To create urls, Go to urls.py of project. We can create urls by using path function
 
        Syntax:
               path(suffix/, function_address, name = 'url_name')
   Note: Create url mapping only within the URL_PATTERNS list variable.

#### Summary
To create and launch a Django project, you can:

    Install Django within a virtual environment
    Create a project using the command django-admin startproject myproject
    Create an app
    Run migrations 

Here are some other things you can do with Django:

    Pin dependencies
    This command writes the names and versions of all external Python packages to a file called requirements.txt. 

* Use a template:
A Django template is an HTML file that contains elements of the Django template language. Templates generate HTML pages based on the context provided by views. 
* Define models:
A model in Django is a Python class that represents a specific database object. 
* Use migrations:
Migrations generate a database schema for your models. They are one of the most used features of Django. 
* Structure projects using apps:
You can structure your projects using apps to keep code for different parts of the project separate. For example, in an e-commerce project, you might have apps called "products", "ordering", and "stock". 

Django is a Python web framework used to develop full-stack web applications
