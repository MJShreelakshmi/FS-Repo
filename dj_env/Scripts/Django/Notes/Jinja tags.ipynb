#### Jinja tagging
Jinja tagging is a modern day tagging (template) language for python developers

        OR
Jinja tagging is an advanced level of writing tags in html files

Jinja tags are also known as Expression tags or Template tags

#### Classification of Jinja tags
Based on the operation we are doing with Jinja tags, they are classified into 2 types:
        
        1. Tags are used for printing the data which is sent from Backend to Frontend
        2. Tags that are used for performing some operations
#### 1. Tags used for printing the data
By using thsus type of tags, we can render some content from Backend and represent the same rendered data in the webpage
#### Representation of Jinja tags which is used for printing data:

            {{ keyname }}
#### Procedure
1. Go to html file, represent jinja tags as shown below

       Syntax:     {{ variable_name }} or {{ key_name }}
       Example:    {{ age }}
2. Go to views.py, create a dictionary inside the function and pass it as value for the context attribute

       Syntax:
                   def function_name( request ):
                       d = {'key': 'value' }
                       return render (request, 'htmlfile_name.ext', context = d )
3. Usage of context attribute name is optional

       Example:
                   def jinja_print(request):
                       d = { 'name': 'Shree', 'age': 20 }
                       return render (request, 'jinja_print.html, context = d )

#### 2. Jinja tags used for performing some operations:
Whenever we want to perform some operations like inserting images, creating hyperlinks etc, we got for this type of tags

#### Representation of Jinja tags for operation"

        {% expression %}
#### Operations that can be done by Jinja tags are as follows
    1. url navigation
    2. if conditional statements
    3. looping statements
    4. insertion of static files like images, CSS, JS etc.
    5. template inheritence
#### 1. Conditional statements by Jinja tags
Syntax for simple if condition:
            
            {% if condition %}
                statements
            {% endif %}

Syntax for simple if-else condition:

            {% if condition %}
                statements of if
            {% else %}
                statements of else
            {% endif %}

Syntax for simple if-elif-else condition:

            {% if condition %}
                statements of if
            {% elif condition %}
                statements of elif 
            {% else %}
                statements of else
            {% endif %}

Syntax for nested-if condition:

            {% if condition %}
                statements of outer if
                {% if condition %}
                    statements of inner if 
                {% endif %}
            {% endif %}

#### 2. Looping statements
1. We cannot use while loop with Jinja tags
2. We cannot use for loop with range along with Jinja tags

Syntax for for-loop with collection data type

        {% for var in CDT %}
            statements of for loop
        {% endfor %}

#### 3. URL Navigation
1. It is the process of travelling from one Html page to another Html page by using url mapping names
2. We can achieve Url Navigation by using Anchor tags

Representation of href value for generic url file:
        
        Generic url navigation is a process of travelling from one Html page to another Html page by using Generic url mapping names
        syntax:   <a href = '{% url "name_of_url" %}'> Content </a>

Representation of href value for specific url file:
        
        Specific url navigation is a process of travelling from one Html page to another Html page by using Specific url mapping names
        syntax:   <a href = '{% url "app_name: name_of_url" %}'> Content </a>

    html files --> templates
    tables     --> models
    css, js, images --> static files

#### Creation of static folders
1. Inorder to store images, js properties and properties, we use static folder
2. As the content inside this will not change, we call this folder as static

We can create a static folder in 2 ways:
1. Generic static folder
2. Specific static folder

#### 1. Generic static folder
In this static folder, all the images, css and js properties of all applications will be stored in one static folder.

#### Steps to create and register static folder:
* Create a project
* cd project
* mkdir static
* cd static
* mkdir css
* mkdir js
* mkdir images
* cd..

#### Registering generic static folder:
* create a variable STATIC_DIR and store the path of static as value

        Eg: STATIC_DIR = os.path.join(BASE_DIR, 'static')

* Create a list variable with name STATICFILES_DIRS and do as shown below:

        Eg: STATICFILES_DIRS = [ STATIC_DIR ]

#### Steps to check wheteher static folders are connected:
* python manage.py runserver
* http://127.0.0.1:8000/static/images/image_name.extesnion
* We need to run the above url, If it is displaying the image, the your static folder is connected succesfully

#### Insertion of Images into Webpages
We use 'img' html tag for inserting images

Syntax for inserting images is django is:

        <img src = "{% static 'images/image_name.extension' %}>
Now we will get an error because we have not loaded static files into the html file

Inorder to avoid this, load the static file into html by using below command

        {% load static %}

#### Connecting CSS to Html:
We us link tag for connecting the css files to html

Syntax for connecting css file in django is:

            <link href="{% static 'css/css_file_name.extension' %}">

#### Specific Static Folders
In any project, CSS and JS properties will be same for all the applications
1. So we will have CSS, JS properties inside inner projects static folder
2. We will have images inside the app static folder

#### Steps to create and register specific static folders

1. Create a project, app
2. Create a static folder inside inner project to store CSS and JS 
3. Create another static folder inside app to store images
4. Go to settings.py file

        1. Register the app
   
        2. Register the templates
   
        3. Create a variable with the name STATIC_DIR_PROJECTNAME to store the path of the project's static folder
            Example: STATIC_DIR_PROJECT1 = =os.path.join(os.path.join(BASE_DIR, 'project1'), 'static')
   
        4. Create a variable woth the name STATIC_DIR_APP to store the path of the app static folder
            Example: STATIC_DIR_APP = =os.path.join(os.path.join(BASE_DIR, 'app'), 'static')
   
        5. Create a list variable STATICFILES_DIR to store the above variables
            Example: STATICFILES_DIR = [STATIC_DIR_PROJECT1, STATIC_DIR_APP ]
   
        6. Inorder to collect and store all the CSS, JS and images from specific static folders, we need to create one root static directory
            Example: STATIC_ROOT = os.path.join(BASE_DIR, 'static')
   
        7. After registering we have to run below command to collect the static files
                        python manage.py collectstatic
Note:
1. Incase of specific static directories, we should not static properties directly into our Root static folder
2. If we make any changes in static folders, it is mandatory to run collectstatic
